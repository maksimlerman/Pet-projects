#### Разделение объектов по типу одноместные-двухместные

<div class="alert alert-warning">
<b>Проблема:</b> Часть объектов, выставленных на продажу являются многоместными парковками и гаражами (чаще всего, двухместными). Иногда об этом пишут в тексте объявления, иногда нет. Двухместные паркинги стоит учитывать по-особому, так как, вероятно, они в два раза дороже (позже мы это проверим). Если учитывать их как одноместные, то можно неправильно сформировать представление об уровне цен в конкретном микрорайоне. </div>


<div class="alert alert-success">
<b>Решение:</b> Разделим все объявления на условно одноместные и двухместные паркинги. Понять, к какому типу паркинга относится объявление можно по площади объекта. Выделим площадь в отдельный столбец из доп.параметров</div>

```python
df_sel['Доп.параметры'] = df_sel['Доп.параметры'].where(df_sel['Доп.параметры'].str.contains('Площадь'), 
                                                        other=np.nan)
df_sel['площадь'] = df_sel['Доп.параметры'].str.split('Площадь=').str[-1]

display(df_sel.head(1))

display(df_sel['площадь'].unique())
```

<div class="alert alert-warning">
<b>Проблема:</b> В некоторых объявлениях площадь не указана, а в некторых указана условно ">30".  </div>

<div class="alert alert-success">
<b>Решение:</b> Там где указано ">30", заменяем значение площади на 30. Не принципиально, сколько конкретно там квадратных метров, важно, что, если больше 30, то такое место или гараж можно отнести к двухместным паркингам. Кроме того, дабавим в датасет тип площади: "площадь определена" и "площадь НЕ определена".</div>

```python
# Там где указано ">30", заменяем значение площади на 30
df_sel['площадь'] = df_sel['площадь'].where(df_sel['площадь'] != '> 30', 30)

# дабавляем параметр "тип площади"
df_sel.loc[:,'вид площади'] = np.select([df_sel.площадь.isna()], 
                                        ['площадь НЕ определена'],
                                        default='площадь определена')


# делим датасет на два - объявления с определенной площадью
df_sel_nan = df_sel.query('площадь.isna()', engine='python')


# смотрим, как распределены по стоимости объекты с определенной и неопределенной площадью
fig, ax = plt.subplots(figsize=(15,6))
sns.kdeplot(data=df_sel, x='Цена', hue='вид площади', ax=ax)
ax.set_title('Распределение объектов по стоиммости с определенной и неопределенной площадью')
ax.text(2000000, 
        0.0000006, 
        f'доля объектов с неопределенной площадью:{round(100*len(df_sel_nan)/len(df_sel), 2)}%', 
        fontsize=16)
plt.show()

df_sel = df_sel.query('~площадь.isna()', engine='python')
```

Несмотря на то, что объектов с неопределенной площадью достаточно мало, около 5%, мы не можем просто отбростить их. Дело в том, что наличие или отсутствие этих объектов мало значимо для всего рынка, но локально в своих микрорайонах учет или неучет этих объектов может дать неправильные ценовые сигналы. Поэтому задача определения, к какому типу относятся эти объекты (одноместным или двухместным) сохраняется.

Проверим, есть ли аномалии в величине площади у тех объектов, площадь которых определена.

```python
df_sel['площадь'] = df_sel['площадь'].astype(float)
print(df_sel['площадь'].describe())
```

Рассмотрим аномально большие по площади объекты

```python
df_sel.query('площадь > 100')
```

```python
# у части объектов площадь указана ошибочно. Судя по описанию, эти объекты относятся к двойным паркингам.
# Поэтому, заменяем их площадь на 30 м2. 
df_sel.loc[369, 'площадь'] = 30
df_sel.loc[1083, 'площадь'] = 30

# остальные объекты с площадью >100м2 считаем аномалиями и удаляем. 
# По смыслу из описания они не относятся к нашему исследованию
df_sel = df_sel.query('площадь <= 100')
```

Теперь в датасет объектов с определенной площадью можно ввести новый признак - признак двойного места. Место считается двойным, если его площадь больше или равна 30 м2. И, если площадь меньше 30 м2, то считаем это одинарным местом.  

```python
df_sel.loc[:,'двойное_место'] = np.select([df_sel.площадь < 30, df_sel.площадь >= 30], 
                                          [0, 1])
```

Посмотрим, как распределены по ценам одноместные паркинги, двухместные паркинги и паркинги с неопределенной площадью.

```python
fig, ax = plt.subplots(figsize=(15,6))
sns.kdeplot(data=df_sel.query('двойное_место == 0'), 
            x='Цена', 
            bw_adjust=.2, 
            ax=ax, 
            label='одноместные места')

sns.kdeplot(data=df_sel.query('двойное_место == 1'), 
            x='Цена', 
            bw_adjust=.2, 
            ax=ax, 
            label='двуместные места')

sns.kdeplot(data=df_sel_nan, 
            x='Цена', 
            bw_adjust=.2, 
            ax=ax, 
            label='площадь не определена')
ax.legend(loc="upper right")
plt.title('Распределение объявлений по стоимости')
plt.show()
```

Как видно из графика, кривая распределения двухместных паркингов смещена вправо, что достаточно очевидно, т.к. чаще всего они дороже одномесных. Кривая распределения объектов, по которым площадь не определена больше походит на кривую распределения одноместных. Однако, из этого нельзя сделать вывод о том, что все объекты с неопределенной площадью являются одноместными паркингами, так как по кривой распределения видно, что среди них есть очень дорогие объекты.

Поэтому выделять из объектов с неопределенной площадью двухместные места нужно другим способом. Попробуем из описания, которые дают собственники выделить слова, которые характерны для одноместных и двухместных объектов.

```python
# сначала задаем перечень наиболее употребляемых слов, 
# которые одинаково часто используются в объявлениях о продаже одноместных и двухместных паркингов
sw = ['продам', 'в', 'на', 'и', 'с', 'по', 'гараж', 'не', 'для', '-',
      'есть', 'под', 'паркинг', 'продается', 'место', 'из', 'паркинге', 'бокс', 'можно', ',',
      'кв.', 'кв.м.', 'за', 'въезд', 'рядом', 'до', 'или', 'жк', 'к', 'все']

def most_common_words(df):
    # Функция most_common_words принимает на вход датафрейм, и определяет ТОП 30 наиболее часто встречающихся слов
    # с разделением по типу объявлений
    description_0 = " ".join(df.query('двойное_место == 0')["Описание"])
    words_0 = [word for word in description_0.split() if word.lower() not in sw]
    description_0 = " ".join(words_0)
    most_common_words_0 = pd.DataFrame(Counter(description_0.split()).most_common(30),
                                        columns = ['одноместные','quantity'])

    description_1 = " ".join(df.query('двойное_место == 1')["Описание"])
    words_1 = [word for word in description_1.split() if word.lower() not in sw]
    description_1 = " ".join(words_1)
    most_common_words_1 = pd.DataFrame(Counter(description_1.split()).most_common(30),
                                        columns = ['двуместные','quantity'])

    return (most_common_words_0
            .join(most_common_words_1, rsuffix='_1')
            .drop(columns = ['quantity','quantity_1'])
            )
most_common_words(df_sel)
```

Нельзя сказать, что данный метод позволил однозначно дифференциировать объекты по количеству мест, но результаты вполне ожидаемые. Например то, что на первом и втором месте по наиболее часто встречающимся словам среди двухместных объектов именно слова "2" и "два". Так же, интересно то, что на третьем месте слово "смотровая". Действительно, смотровые ямы чаще свойственны большим гаражам. (Для одноместных объектов "смотровая" только на 27-м месте)

## Поиск инвестиционных объектов 


Задача определения объектов наиболее интересных для инвестирования заключается в следующем. Среди всех выставленных на родажу объектов нужно найти такие, которые расположены в зонах повышенного арендного спроса. Т.е. там, где сформированы наибольшие арендные ставки. И если соотношение цены продающегося объекта и средней (или медианной) цены на аренду в этой зоне находится на определенном уровне, обеспечивающим возврат инвестиций в заданные сроки, то такой объект является перспективным с т.з. инвестирования для сдачи его в аренду.


Для решения этой задачи необходимо сформировать зоны (кластеры) по территориальному признаку. Для этого используем широту и долготу расположения объектов. Разбиение на кластеры осуществляем алгоритмом KMeans.

### Определение арендных кластеров

```python
# обязательная стандартизация данных перед работой с алгоритмами
sc = StandardScaler()
X_sc = sc.fit_transform(df_rent_single_slot[['lat','lng']])
```

Подбираем число кластеров, которое обеспечит наиболее эффективное разбиение. Эффективность определяем по метрике silhouette_score.

В качестве обучающей модели используем метод К ближайших соседий KNeighborsClassifier. И для начала, подбираем налучший параметр n_neighbors - то самое число соседий, по которому модель определяет, к какому кластеру должен относиться объект. Поиск налучшего значения можно осуществить с помощью класса GridSearchCV – поиск наилучшего набора параметров, доставляющих минимум ошибке перекрестного контроля.

Проверим, чему равна точность на тестовой выборке при этом значении параметра

```python
warnings.simplefilter('ignore')
number_of_clusters=[]
slhtte_score = []
bce = []
bnn = []
r2 = []
accuracy = []
for n in range(50, 550, 50):
    km = KMeans(n_clusters=n, random_state=0) # задаём число кластеров, равное n, 
                                              # и фиксируем значение random_state для воспроизводимости результата
    labels = km.fit_predict(X_sc) # применяем алгоритм к данным и формируем вектор кластеров 
    
    number_of_clusters.append(n)
    slhtte_score.append(silhouette_score(X_sc, labels))
    
    # сохраняем метки кластера в поле нашего датасета
    df_rent_single_slot['rent_cluster'] = labels
    
    # создаем обучающую и валидационную выборки
    X_train, X_test, y_train, y_test = train_test_split(df_rent_single_slot[['lat','lng']], 
                                                    df_rent_single_slot['rent_cluster'], 
                                                    test_size=0.2)

    # задаем возможные варианты числа n_neighbors
    n_neighbors_array = [1, 3, 5, 7, 10, 15]
    knn = KNeighborsClassifier()
    grid = GridSearchCV(knn, param_grid={'n_neighbors': n_neighbors_array})
    grid.fit(X_train, y_train)
    
    #  поиск наилучшего набора параметров, доставляющих минимум ошибке перекрестного контроля.
    best_cv_err = 1 - grid.best_score_
    best_n_neighbors = grid.best_estimator_.n_neighbors
    
    bce.append(best_cv_err)
    bnn.append(best_n_neighbors)
    
    # обучаем модель
    knn = KNeighborsClassifier(n_neighbors=best_n_neighbors)
    knn.fit(X_train, y_train)
    
    # опробуем модель на тестовой выборке
    validates = knn.predict(X_test)
    
    # оцениваем результат
    r2.append(r2_score(y_test, validates))
    accuracy.append(accuracy_score(y_test, validates))

    
k_means_results = pd.DataFrame({'number_of_clusters':number_of_clusters,
                                'silhouette_score':slhtte_score,
                                'best_cv_err':bce,
                                'best_n_neighbors':bnn,
                                'r2':r2,
                                'accuracy':accuracy})

k_means_results
```

```python
k_means_results.plot(x='number_of_clusters',
                     y=['silhouette_score', 'accuracy'],
                     figsize=(15,6),
                     title='Эффетивность разбиения на кластеры')
```

Наиболее эффективное разбиение на кластеры достигается при количестве кластеров равном 150. Точность и значение silhouette_score при этом значении практически максимальные.

```python
warnings.simplefilter('ignore')
km = KMeans(n_clusters=150, random_state=0) # задаём число кластеров, равное 150, 
                                          # и фиксируем значение random_state для воспроизводимости результата
labels = km.fit_predict(X_sc) # применяем алгоритм к данным и формируем вектор кластеров 

# сохраняем метки кластера в поле нашего датасета
df_rent_single_slot['rent_cluster'] = labels
```

### Создание и обучение модели - классификатора

Теперь, когда у нас определены кластеры, создадим модель, которая по каординатам объекта сможет определить, к какому кластеру относится этот объект.

В качестве обучающей модели используем метод К ближайших соседий KNeighborsClassifier. Параметр n_neighbors принимаем равным 1 - именно это значение обеспечивает наименьшую ошибку перекрестного контроля.

Проверим, чему равна ошибка на тестовой выборке при этом значении параметра

```python
# создаем обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(df_rent_single_slot[['lat','lng']], 
                                                    df_rent_single_slot['rent_cluster'], 
                                                    test_size=0.2)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
validates = knn.predict(X_test)

print('R2: {:.2f}'.format(r2_score(y_test, validates)))
print('Точность: {:.2f}'.format(accuracy_score(y_test, validates)))
```

Определяем для каждого объекта, выставленного на продажу срок окупаемости в месяцах payback. Срок окупаемости находится делением стоимости объекта на среднюю арендную ставку кластера.

Теперь, когда мы знаем распределение объектов на аренду по кластерам, составляем таблицу с характеристикой кластеров. Нас интересует инвестиционная привлекательность кластера. Примем, что кластер тем привлекательнее, чем выше в нем стоимость аренды и чем меньше в нем срок арендной экспозиции. Это следует из предположения о том, что если в определенном кластере сложилась высокая цена аренды, а объекты сдаются быстро, то в этом кластере есть повышенный спрос на аренду паркингов.

```python
# для каждого кластера вычисляем среднюю стоимость аренды и медианный срок экспозиции объявлений
cluster_balance = pd.DataFrame(df_rent_single_slot
                               .groupby('rent_cluster')
                               .agg({'Цена':'mean',
                                     'expo':'median'}))
cluster_balance = cluster_balance.rename(columns={'Цена':'mean_rent_price',
                                                  'expo':'median_expo_period'})


# вычисляем рейтинг кластера. Медианную стоимость аренды и медианный срок экспозиции ранжируются от 0 до 1. 
# Где 1-максимальная стоимость аренды, 0 - минимальная
# Где 1-минимальный срок экспозиции, 0 - максимальный
# показатели суммируются и образуют рейтинг
cluster_balance['cluster_rating'] =\
(cluster_balance['mean_rent_price'] / cluster_balance['mean_rent_price'].max())+\
(1-(cluster_balance['median_expo_period'] / cluster_balance['median_expo_period'].max()))

# получим список лучших кластеров. Исходя из того, что рейтинг кластера лежит в диапазоне от 0 до 2,
# к лучшим отнесем те, чей рейтинг выше 1
best_clusters = cluster_balance.query('cluster_rating > 1').index
best_clusters
```

```python
# добавим рейтинг кластера в таблицу с объектами
df_sel_single_slot = df_sel_single_slot.merge(cluster_balance.cluster_rating, on='rent_cluster')
df_sel_duble_slot = df_sel_duble_slot.merge(cluster_balance.cluster_rating, on='rent_cluster')
```
